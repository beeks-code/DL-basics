```python
import gensim.downloader as api
model=api.load("word2vec-google-news-300")
```

```python
### semantic embedding for normal word
model["hello"]
```

```python
# model["king"]-model["man"]+model["women"]
model.most_similar(positive=["king","women"],negative=["man"])
```

```python
### similarity between words
print(model.similarity("men","women"))
print(model.similarity("deer","tiger"))
print(model.similarity("car","bike"))
print(model.similarity("sun","men")) # random
print(model.similarity("father","mother"))
```

```python
### similar to given word
model.most_similar("Messi",topn=4)
```
