```python
import pandas as pd
```

```python
df=pd.read_csv(r"hawa/NABIL.csv")
```

```python
df.tail(3)
```

```python
df["Volume"]=df["traded_amount"]/df["traded_quantity"]
df["average"]=(df["close"]+df["high"])/2
df["Next_Open"]=df["open"].shift(-1)
df["average"]=df["average"].shift(-1)
```

```python
df=df.dropna()
```

```python
df.head(2)
df["Next_Open"]=df["open"].shift(-1)
```

```python
df=df.dropna()
```

```python
df.tail(3)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
input_df=df.iloc[:,1:-2]
output_df=df.iloc[:,[-1,-2]]
```

```python
input_df.tail(4)
```

```python
output_df.head(3)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled=scaler.fit_transform(input_df,output_df)
```

```python
model=Sequential()
model.add(Dense(4,activation="relu",input_shape=(9,)))
model.add(Dense(124,activation="relu"))
model.add(Dense(2))
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',loss='mse', metrics=['mae']  
)
```

```python
model.fit(X_scaled,output_df, epochs=50, batch_size=32, validation_split=0.2)
```
