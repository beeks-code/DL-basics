Positional Encoding used in "Attention is all you need paper"

$$
\begin{aligned}
PE(pos, 2i) &= \sin\left(\frac{pos}{10000^{2i/d}}\right) \\
PE(pos, 2i+1) &= \cos\left(\frac{pos}{10000^{2i/d}}\right)
\end{aligned}
$$


```python
import numpy as np
```

```python
d=6
```

```python
def PE1(pos: int, i: int, d: int):
    pwr = 2 * i / d
    return np.sin(pos / (10000 ** pwr))
```

```python
def PE2(pos: int, i: int, d: int):
    pwr = 2 * i / d
    return np.cos(pos / (10000 ** pwr))
```

```python
def get_embedding(pos:int,d:int):
    emd=[] ## embeding array
    for i in range(d//2):
        emd.append(PE1(pos,i,d)) ## sin function
        emd.append(PE2(pos,i,d))  ## cosine function

    return emd    
```
